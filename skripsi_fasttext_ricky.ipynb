```python
import os
import fasttext
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load model FastText yang sudah dilatih
model_path = 'fasttext_model.bin'
model = fasttext.load_model(model_path)

# Lokasi direktori di mana file-file ini berada
directory_path = "/content"

# Daftar file yang ingin Anda cek
files_to_check = [f"userSB{i:02d}.txt" for i in range(1, 30)]

# File yang ingin diukur semantik similarity-nya dengan file TeacherMap.txt
target_file_names = ["userSB04.txt", "userSB18.txt"]

# Membaca konten file TeacherMap.txt
teacher_map_file_path = os.path.join(directory_path, "TeacherMap.txt")
with open(teacher_map_file_path, 'r') as teacher_map_file:
    cleaned_teacher_map_content = teacher_map_file.read().replace('\n', ' ')

# Mengukur kemiripan semantik antara file TeacherMap.txt dan file userSB01.txt sampai userSB29.txt
semantic_similarity_results = {}

for file_name in files_to_check:
    if file_name in target_file_names:
        continue  # Skip file yang dikecualikan

    file_path = os.path.join(directory_path, file_name)

    try:
        with open(file_path, 'r') as file:
            content = file.read().replace('\n', ' ')
            # Menghitung cosine similarity menggunakan sklearn
            similarity_score = cosine_similarity([model.get_sentence_vector(cleaned_teacher_map_content)],
                                                [model.get_sentence_vector(content)])[0][0]
            semantic_similarity_results[(file_name, "TeacherMap.txt")] = similarity_score
    except FileNotFoundError:
        print(f"File {file_name} tidak ditemukan")

# Menampilkan hasil kemiripan semantik
for pair, similarity in semantic_similarity_results.items():
    print(f"Cosine similarity between {pair[0]} and {pair[1]}: {similarity:.4f}")
```

```python
import matplotlib.pyplot as plt

# Menyusun hasil kemiripan menjadi daftar similarities
similarities = []

for pair, similarity in semantic_similarity_results.items():
    similarities.append((pair[0], pair[1], similarity))

# Mengurutkan similarities berdasarkan nilai similarity (dari terbesar ke terkecil)
similarities.sort(key=lambda x: x[2], reverse=True)

# Membuat grafik
file_pairs = [f"{pair[0]} - {pair[1]}" for pair in similarities]
sorted_similarities = [pair[2] for pair in similarities]

plt.figure(figsize=(12, 8))
plt.barh(file_pairs, sorted_similarities, color='dodgerblue')
plt.xlabel('Cosine Similarity')
plt.title('Semantic Similarity between Files and TeacherMap (Sorted by Similarity)')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import jaccard_score
import os

# Lokasi direktori di mana file-file ini berada
directory_path = "/content"

# Daftar file yang ingin Anda cek
files_to_check = [f"userSB{i:02d}.txt" for i in range(1, 30)]

# File yang ingin diukur Jaccard similarity-nya dengan file TeacherMap.txt
target_file_names = ["userSB04.txt", "userSB18.txt"]

# Membaca konten file TeacherMap.txt
teacher_map_file_path = os.path.join(directory_path, "TeacherMap.txt")
with open(teacher_map_file_path, 'r') as teacher_map_file:
    cleaned_teacher_map_content = teacher_map_file.read().replace('\n', ' ')

# Membuat CountVectorizer untuk mengubah teks menjadi representasi bag-of-words
vectorizer = CountVectorizer()

# Menghitung Jaccard similarity antara file TeacherMap.txt dan file userSB01.txt sampai userSB29.txt
jaccard_similarities = []
for file_name in files_to_check:
    if file_name in target_file_names:
        continue  # Skip file yang dikecualikan

    file_path = os.path.join(directory_path, file_name)

    try:
        with open(file_path, 'r') as file:
            content = file.read().replace('\n', ' ')

            # Mengubah teks menjadi representasi bag-of-words
            X = vectorizer.fit_transform([cleaned_teacher_map_content, content])

            # Menghitung Jaccard similarity
            jaccard_similarity = jaccard_score(X[0].toarray()[0], X[1].toarray()[0], average='weighted')
            jaccard_similarities.append((file_name, "TeacherMap.txt", jaccard_similarity))
    except FileNotFoundError:
        print(f"File {file_name} tidak ditemukan")

# Menampilkan hasil Jaccard similarity
for pair in jaccard_similarities:
    print(f"Jaccard similarity between {pair[0]} and {pair[1]}: {pair[2]:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Mengurutkan similarities berdasarkan nilai similarity (dari terbesar ke terkecil)
jaccard_similarities.sort(key=lambda x: x[2], reverse=True)

# Membuat grafik
file_names = [f"{pair[0]} - {pair[1]}" for pair in jaccard_similarities]
sorted_jaccard_similarities = [pair[2] for pair in jaccard_similarities]

plt.figure(figsize=(10, 8))
bars = plt.barh(file_names, sorted_jaccard_similarities, color='dodgerblue')
plt.xlabel('Jaccard Similarity')
plt.title('Similarity between Files (Jaccard Similarity)')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Menambahkan nilai Jaccard similarity pada setiap bar
for bar, similarity in zip(bars, sorted_jaccard_similarities):
    plt.text(bar.get_width() - 0.02, bar.get_y() + bar.get_height() / 2, f'{similarity:.3f}',
             va='center', ha='right', color='black')

plt.tight_layout()
plt.show()
```
